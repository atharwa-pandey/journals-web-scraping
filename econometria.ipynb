```python
def make_directory(Volume,Issue):
    directory = 'Volume ' + str(Volume) + ' Issue ' + str(Issue) + ' Year ' + str(Volume+1932)
    # Parent Directory path 
    parent_dir = "D:/Results/Econometria"
    # Path 
    path = os.path.join(parent_dir, directory) 
  
    # Create the directory 
    # 'Nikhil' 
    try: 
        os.makedirs(path, exist_ok = True) 
        print("Directory '%s' created successfully" % directory) 
    except OSError as error: 
        print("Directory '%s' can not be created" % directory) 
        
    os.chdir(path)
    print(os.getcwd())
```

```python
def find_table2(url2):
    print(url2)
    r.init()
    r.url(url2)
    r.wait(random.randrange(5,10,1))
    html2 = r.read('page')
    r.close()
    soup = BeautifulSoup(html2, 'html.parser')
    return soup.find('div', attrs={'id': 'article__content'})
```

```python
def download_pdf(pdf_url):
    try:
        r.init()
        r.url(pdf_url)
        r.wait(random.randrange(5,10,1))
        #pdf_url_main = r.url()
        #r.download(download_url = pdf_url_main , filename_to_save = journal_name+".pdf")
        r.close()
                
    except:
        r.close()
        download_pdf(pdf_url)
```
